### Konsolidacja??

```python
def konsoliduj():
    result = answers.loc[:,['Wiek', 'Płeć', 'id_response']]
    
    for i in range(3,31,4):
        nazwa = answers.columns[i]
        
        wartosc = answers.iloc[:,i:(i+4)].mean(axis=1)
        std = answers.iloc[:,i:(i+4)].std(axis=1)
        
        result[nazwa+' wartosc'] = wartosc
        result[nazwa+' std'] = std
        
    return result
```

```python
answers_konsolidacja = konsoliduj()
```

# EDA

## answers

```python
stdsc = StandardScaler()
answers_std = stdsc.fit_transform(answers.drop(columns='id_response'))
```

```python
tsne = MulticoreTSNE(n_jobs=10)
tsne.fit(answers_std)
```

```python
pca = PCA(n_components=2)
answers_pca = pca.fit_transform(answers_std)
```

### Podział zbioru względem wieku - równe przedziały

```python
def podzial_zbioru_szerokosciowo(zbior, tsne, pca, zmienna, przedzialy):
    cut, bins = pd.cut(zbior.loc[:,zmienna], bins=przedzialy, labels=np.arange(przedzialy), retbins=True)

    sns.scatterplot(tsne.embedding_[:,0],tsne.embedding_[:,1], s=500, alpha=0.8,
                    hue=np.array(cut), palette=sns.color_palette('bright',przedzialy))
    plt.title('TSNE ' + zmienna, size=50)
    plt.show()
    
    sns.scatterplot(pca[:,0],pca[:,1], s=500, alpha=0.8,
                    hue=np.array(cut), palette=sns.color_palette('bright',przedzialy))
    plt.title('PCA ' + zmienna, size=50)
    plt.show()
```

```python
podzial_zbioru_szerokosciowo(answers, tsne, answers_pca, 'Wiek', 2)
```

### Podział zbioru względem wieku - równe liczności

```python
def podzial_zbioru_ilosciowo(zbior, tsne, pca, zmienna, przedzialy):
    bins = np.linspace(0,1, przedzialy + 1)
    bins = zbior.loc[:,zmienna].quantile(bins)
    cut= pd.cut(zbior.loc[:,zmienna], bins=bins)

    sns.scatterplot(tsne.embedding_[:,0],tsne.embedding_[:,1], s=500, alpha=0.8,
                    hue=np.array(cut), palette=sns.color_palette('bright',przedzialy))
    plt.title('TSNE ' + zmienna,size=50)
    plt.show()
    
    sns.scatterplot(pca[:,0],pca[:,1], s=500, alpha=0.8,
                    hue=np.array(cut), palette=sns.color_palette('bright',przedzialy))
    plt.title('PCA ' + zmienna, size=50)
    plt.show()
```

```python
podzial_zbioru_ilosciowo(answers, tsne, answers_pca, 'Wiek',5)
```
